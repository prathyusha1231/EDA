```python
features_not_used = ['Zeitstempel',
                     'Yearly brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country',
                     'Yearly stocks one year ago. Only answer if staying in same country', 
                     'Yearly bonus one year ago. Only answer if staying in same country', '0']
len(features_not_used)
```

```python
features_cat = ['Gender', 'City', 'Seniority level',
       'Position (without seniority)', 'Years of experience',
       'Your main technology / programming language',
       'Main language at work', 'Company name ', 'Company size',
       'Company type', 'Сontract duration', 'Company business sector']
len(features_cat)
```

```python
features_num = ['Age',
                'Yearly brutto salary (without bonus and stocks)',
                'Yearly bonus','Yearly stocks','Number of vacation days', 
                'Number of home office days per month']
len(features_num)
```

```python
len(features_cat+ features_num+ features_not_used) - len(data.columns)
```

```python
# fill missing values

# categorical features
missing_text = 'MISSING'
data[features_cat] = data[features_cat].fillna(missing_text)

# numerical features
missing_num = -1
data[features_num] = data[features_num].fillna(missing_num)
```

```python
# numerical features
missing_num = -1
data[features_num] = data[features_num].fillna(missing_num)
```

```python
temp_count = data['Position (without seniority)'].value_counts()
temp_count
```

```python
freq_min = 3
current_feature = 'Position (without seniority)'
keep_levels = list(temp_count[temp_count.values>=freq_min].index)
data[current_feature] = data[current_feature].where(data[current_feature].isin(keep_levels), '_OTHER_')
print('\nREDUCED TO:')
print(data[current_feature].value_counts())
```

```python
data['AgeGroup'] = pd.cut(data.Age, bins=[-2,0,20,25,30,35,40,45,50,60,70])
data.AgeGroup.value_counts().sort_index().plot(kind='bar')
plt.title('Age Groups')
plt.grid()
plt.show()
```

```python
for f in features_cat:
    plt.figure(figsize=(14,5))
    val_c = data[f].value_counts()
    if len(val_c) <= 20:
        val_c.plot(kind='bar', color='orange')
        plt.title(f)
    else: # if more than 20 levels show only the most frequent 20
        val_c[0:20].plot(kind='bar', color='orange')
        plt.title(f + ' - Top 20 only')
        
    plt.grid()
    plt.show()
```

```python
for f in features_num:
    plt.figure(figsize=(14,5))
    data[f].plot(kind='hist', bins = 50, color='orange')
    plt.title(f)
    plt.grid()
    plt.show()
```
